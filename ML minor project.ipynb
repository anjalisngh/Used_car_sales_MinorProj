```python
#changing marks to grades
df['marks_total'] = df['math score']+df['reading score']+df['writing score']
df['percentage'] = df['marks_total']/3
'''
Lets assign grades.

Criteria of the grades are as follows:

    90-100 : Grade A
    80-89  : Grade B
    70-79  : Grade C
    60-69  : Grade D
    <69    : Grade F
'''
#method for assigning the grades

def determine_grade(scores):
    if scores >= 90 and scores <= 100:
        return 'Grade A'
    elif scores >= 80 and scores < 90:
        return 'Grade B'
    elif scores >= 70 and scores < 80:
        return 'Grade C'
    elif scores >= 60 and scores < 70:
        return 'Grade D'
    elif scores >= 0 and scores < 60:
        return 'Grade E'
#Calling the method
df['grades']=df['percentage'].apply(determine_grade)
```

```python
#pie chart representation for % of studentss in each grade
df['grades'].value_counts().plot.pie(autopct="%1.2f%%")
plt.title('percentage of students in each grade')
plt.show()
```
